```
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_moons
from sklearn.tree import DecisionTreeClassifier
sns.set(style='darkgrid')
```

Generowanie danych

```
raw_data = make_moons(n_samples=700, noise=0.25)
raw_data
```

```
data = raw_data[0]
data
```

```
target = raw_data[1]
target
```

```
plt.scatter(data[:, 0], data[:,1], c=target, cmap='viridis')
```

Podzia na zbir treningowy i testowy 

```
from sklearn.model_selection import train_test_split

train_test_split?
```

```
X_train, X_test, y_train, y_test = train_test_split(data, target)
```

```
print('X_train', X_train.shape)
print('X_test', X_test.shape)
print('Y_train', y_train.shape)
print('Y_test', y_test.shape)
```

```
plt.figure(figsize=(8,6))
plt.scatter(X_train[:, 0], X_train[:,1], c=y_train, cmap='RdYlBu', label='training_set')
plt.scatter(X_test[:, 0], X_test[:,1], c=y_test, cmap='RdYlBu', label='test_set', marker='x')
plt.legend()
```

Niedouczenie model regresji logistycznej 

```
from sklearn.linear_model import LogisticRegression
from mlxtend.plotting import plot_decision_regions
```

```
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)

plot_decision_regions(X_train, y_train, log_reg)
plt.title('Dokadno na zbiorze treningowym: {:4f}'.format(log_reg.score(X_train, y_train)))
```

```
plot_decision_regions(X_test, y_test, log_reg)
plt.title('Dokadno na zbiorze treningowym: {:4f}'.format(log_reg.score(X_train, y_train)))
```

```
tree = DecisionTreeClassifier(max_depth = 2)
tree.fit(X_train, y_train)
```

```
plot_decision_regions(X_train, y_train, tree)
plt.title('Dokadno na zbiorze treningowym: {:4f}'.format(tree.score(X_train, y_train)))
```

```
plot_decision_regions(X_test, y_test, tree)
plt.title('Dokadno na zbiorze testowy: {:4f}'.format(tree.score(X_test, y_test)))
```

Grid Search
