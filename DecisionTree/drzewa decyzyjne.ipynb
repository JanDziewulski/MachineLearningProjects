```python
def make_decision_tree(max_depth=1):
    tree_clf = DecisionTreeClassifier(criterion ='gini', max_depth = max_depth, random_state=2 )
    tree_clf.fit(X,y)
    dot_data = StringIO()
    export_graphviz(tree_clf,
                  out_file=dot_data,
                  feature_names= feature_names[:-2],
                  class_names=target_names,
                  rounded=True,
                  filled=True
                )
    graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
    graph.write_png('graph.png')
    plot_decision_regions(X, y, tree_clf)


    plt.xlabel('petal length (cm)')
    plt.ylabel('petal wigth (cm)')
    plt.title('Drzewo decyzyjne: max_depth=1')

    print('Dok≈Çadno≈õc modelu: ' +  str(round(tree_clf.score(X,y) * 100, 2)) + '%')

    return Image(graph.create_png())
```

###Information gain

```python
def entropia(x):
    return -np.sum(x*np.log2(x))
```

```python
entropia([0.5,0.5])
```

```python
entropia([0.8,0.2])
```

```python
p = np.arange(0.1, 1.0, 0.01)
q = 1 - p
pq = np.c_[p,q]
```

```python
entropie = [entropia(pair) for pair in pq]
```

```python
print(entropie)
```

```python
plt.plot(p, entropie)
```
